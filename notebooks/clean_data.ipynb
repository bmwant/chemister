```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
filename = 'data/uah_to_rub_2017.csv'
```

```python
df = pd.read_csv(filename)
```

```python
df
```

```python
df.describe()
```

```python
df.drop_duplicates(subset=['date'], inplace=True)
```

```python
df['buy_nb'] == df['sale_nb']  # same values
```

```python
df.rename({'buy_nb': 'nb'}, axis='columns', inplace=True)
```

```python
df.columns
```

```python
df.drop(columns=['sale_nb'], inplace=True)
```

```python
df.describe()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')
```

```python
df.sort_values(by=['date'], inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
prev_row = df.loc[df['date'] == '23.08.2017']
next_row = df.loc[df['date'] == '25.08.2017']
prev_row[['buy', 'sale', 'nb']].reset_index(drop=True) == next_row[['buy', 'sale', 'nb']].reset_index(drop=True)
```

```python
df.loc[df['date'] == '24.08.2017'] = prev_row
```

```python
copy_prev = df.loc[df['date'] == '25.08.2017']
copy_prev.loc[copy_prev['date'] == '25.08.2017']['date'] = '24.08.2017'
copy_prev.reset_index(drop=True, inplace=True)
copy_prev
```

```python
df1 = pd.concat([df, copy_prev], ignore_index=True)
```

```python
df1.sort_values(by=['date'], inplace=True)
```

```python
df1.describe()
```

```python
df1.reset_index(drop=True, inplace=True)
```

```python
df1.set_value(236, 'date', '24.08.2017')
```

```python
df1.sort_values(by=['date'], inplace=True)
```

```python
df1.describe()
```

```python
df1.loc[df1['date'] == '24.08.2017']
```

```python
# NOTE: make sure your store enough digits after point, for some currencies it might be critical
df1 = df1.round({'buy': 2, 'sale': 2, 'nb': 6})
```

```python
df1.describe()
```

```python
df1.to_csv(filename, date_format='%d.%m.%Y', index=False)
```
