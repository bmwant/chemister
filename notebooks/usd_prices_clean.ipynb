```python
# Reorder csv file and clean unused columns
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
filename = 'data/uah_to_usd_2015.csv'
```

```python
df = pd.read_csv(filename)
```

```python
df
```

```python
df.describe()
```

```python
df['buy_nb'] == df['sale_nb']  # same values
```

```python
df.rename({'buy_nb': 'nb'}, axis='columns', inplace=True)
```

```python
df.columns
```

```python
df.drop(columns=['sale_nb'], inplace=True)
```

```python
df.describe()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')
```

```python
df.sort_values(by=['date'], inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df = df.round({'buy': 2, 'sale': 2, 'nb': 6})
```

```python
df
```

```python
df.to_csv(filename, date_format='%d.%m.%Y', index=False)
```
